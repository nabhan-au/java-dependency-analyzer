```python
def compare_artifact_with_tag(artifact, versions, tags):
    result = []
    if (versions == [] or tags == [] or versions is None or tags is None or len(versions) == 0 or len(tags) == 0):
        return result
    version_list = ast.literal_eval(versions)
    tag_list = ast.literal_eval(tags)
    try:
        for version in version_list:
            isFound = False
            for tag in tag_list:
                if version == tag:
                    result.append([version, tag])
                    isFound = True
            for tag in tag_list:
                if version in tag and artifact in tag and not isFound:
                    result.append([version, tag])
                    isFound = True
            for tag in tag_list:
                if version in tag and len(tag) - len(version) <= 2  and not isFound:
                    result.append([version, tag])
                    isFound = True
            if not isFound:
                result.append([version, None])
    except:
        print(versions, tags)
    return result
```

```python
df = df.fillna("")
df["compared_result"] = None
count_index = 0
for index, row in df.iterrows():
    result = compare_artifact_with_tag(row["artifact_id"].split(":")[-1], row["versions"], row["tags"])
    df.at[index, "compared_result"] = result
    count_index += 1
    
# df.drop("tags", axis=1)    
# df.drop("artifact_id", axis=1, inplace=True)
```
