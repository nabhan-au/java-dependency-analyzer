```python
import pandas as pd
import ast
```

```python
df = pd.read_csv("data/success-project-with-version-tag.csv")
```

```python
def compare_artifact_with_tag(artifact, ver_dict, tag_dict):
    result = []
    
    version_list = []
    tag_list = []
    parsed_ver_dict = ast.literal_eval(ver_dict)
    parsed_tag_dict = ast.literal_eval(tag_dict)
    for ver in parsed_ver_dict:
        version_list.append(ver["name"])
    for tag in parsed_tag_dict:
        tag_list.append(tag["name"])
    if (version_list == [] or tag_list == [] or len(version_list) == 0 or len(tag_list) == 0):
        return result
    try:
        for version in version_list:
            isFound = False
            for tag in tag_list:
                if version == tag:
                    result.append([version, tag])
                    isFound = True
            for tag in tag_list:
                if version in tag and artifact in tag and not isFound:
                    result.append([version, tag])
                    isFound = True
            for tag in tag_list:
                if version in tag and len(tag) - len(version) <= 2  and not isFound:
                    result.append([version, tag])
                    isFound = True
            if not isFound:
                result.append([version, None])
    except:
        print(ver_dict, tag_dict)
    return result
```

```python
df = df.fillna("")
df["compared_result"] = None
count_index = 0
for index, row in df.iterrows():
    result = compare_artifact_with_tag(row["artifact_id"].split(":")[-1], row["versions"], row["tags"])
    df.at[index, "compared_result"] = result
    count_index += 1
```
